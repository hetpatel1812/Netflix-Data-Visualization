```python
#import the libraries
import pandas as pd
import matplotlib.pyplot as plt

#load the data
df = pd.read_csv("/netflix_titles.csv")

#clean the data
df=df.dropna(subset=['type','release_year','rating','country','duration'])
type_counts=df['type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_counts.index,type_counts.values,color=['blue','green'])
plt.title("Number of Movies and TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

```

```python
rating_counts=df['rating'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(rating_counts.values,labels=rating_counts.index,autopct="%1.1f%%",startangle=90)
plt.title("Distribution of Ratings on Netflix")
plt.tight_layout()
plt.show()


```

```python
release_counts=df['release_year'].value_counts().sort_index()
plt.figure(figsize=(10,6))
plt.scatter(release_counts.index,release_counts.values,color="red")
plt.title("Number of Movies and TV Shows Released Each Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
country_counts=df['country'].value_counts().head()
plt.figure(figsize=(8,6))
plt.barh(country_counts.index,country_counts.values,color="lightgreen")
plt.title("Top 5 Countries with the Most Movies and TV Shows on Netflix")
plt.xlabel("Count")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
content_by_year = df.groupby(['release_year','type']).size().unstack().fillna(0)
fig, ax =plt.subplots(1,2,figsize=(12,6))

ax[0].plot(content_by_year.index,content_by_year['Movie'],marker='o',linestyle='-',color='blue')
ax[0].set_title('Movie Count by Year')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Number of movies')

ax[1].plot(content_by_year.index,content_by_year['TV Show'],marker='o',linestyle='-',color='orange')
ax[1].set_title('TV Show Count by Year')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Number of TV Shows')

fig.suptitle('Comparison of Movies and TV Shows Released per year')
plt.tight_layout()
plt.show()
```
